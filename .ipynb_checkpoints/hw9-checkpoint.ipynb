```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.norm(loc=21, scale=4)
sample = rv.rvs(size=1000)
rv_2 = sts.expon(scale=(1/(10/60)))
sample_2 = rv_2.rvs(size=1000)
```

```python
x = np.linspace(0,35,1000)
pdf = rv.pdf(x)
plt.figure(figsize=[18, 8])
plt.hist(sample, density=True, bins=30, alpha=0.4, label='sample histogram (normal distribution)')
plt.hist(sample_2, density=True, bins=30, label='sample histogram (exponential distribution)')
plt.plot(x, pdf, label='theoretical pdf', alpha=2)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
import random
#параметры запуска: размеры выборок и количество повторов генерации
#n1 = random.randint(2, 500)
sample_count = []
samples_count = 1000
#массивы для сохранения средних по выборкам
a1 = np.array([])
a2 = np.array([])
a1_exp = np.array([])
a2_exp = np.array([])
for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    n1 = random.randint(2, 50)
    n2 = random.randint(2, 500)
    sample1 = rv.rvs(size=n1)
    sample2 = rv.rvs(size=n2)
    sample1_exp = rv_2.rvs(size=n1)
    sample2_exp = rv_2.rvs(size=n2)
    sample_count.append(n1)
    # считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    a2 = np.append(a2, sample2.mean())
    a1_exp = np.append(a1_exp, sample1_exp.mean())
    a2_exp = np.append(a2_exp, sample2_exp.mean())
```

```python
sum(sample_count)/len(sample_count)
```

```python
print('scenario 1 samples mean for norm. distrib. =', '{:,.2f}'.format(a1.mean()).replace(',',' '))
print('scenario 2 samples mean for norm. distrib. =', '{:,.2f}'.format(a2.mean()).replace(',',' '))
print('scenario 1 samples mean for exp. distrib. =', '{:,.2f}'.format(a1_exp.mean()).replace(',',' '))
print('scenario 2 samples mean for exp. distrib. =', '{:,.2f}'.format(a2_exp.mean()).replace(',',' '))
```

```python
#строим графики
#подписи для графиков
plt.figure(figsize=[18, 8])
label_1 = 'sample of ' + str(samples_count) + ' histogram. scenario 1 for normal distribution with max(n)=50'
label_2 = 'sample of ' + str(samples_count) + ' histogram. scenario 2 for normal distribution with max(n)=500'
label_3 = 'sample of ' + str(samples_count) + ' histogram. scenario 1 for exponential distribution with max(n)=50'
label_4 = 'sample of ' + str(samples_count) + ' histogram. scenario 2 for exponential distribution with max(n)=500'
plt.hist(a1, density=True, label=label_1, bins=100, alpha=0.4)
plt.hist(a2, density=True, label=label_2, bins=100, alpha=0.4)
plt.hist(a1_exp, density=True, label=label_3, bins=100, alpha=0.4)
plt.hist(a2_exp, density=True, label=label_4, bins=100, alpha=0.4)
plt.legend()

plt.xlabel('samples means')
plt.show()
```
